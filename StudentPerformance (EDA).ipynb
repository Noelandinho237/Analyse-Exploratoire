## CATEGORISATION DES NOTES EN 5 
1. 0-35  Failed 
2. 35-49 Fair
3. 50-64 Passed
4. 66-80 Good
5. 81-90 Very Good
6. 90-100 Excellent

```python
data.describe()
```

```python
# Conversion de writing score
data['writing score categ'] = 'Failed'

data.loc[data['writing score'] <= 35, 'writing score categ'] = 'Failed'
data.loc[(data['writing score'] > 35) & (data['writing score'] <= 49), 'writing score categ'] = 'Fair'
data.loc[(data['writing score'] >= 50) & (data['writing score'] <= 64), 'writing score categ'] = 'Passed'
data.loc[(data['writing score'] >= 66) & (data['writing score'] <= 80), 'writing score categ'] = 'Good'
data.loc[(data['writing score'] >= 81) & (data['writing score'] <= 90), 'writing score categ'] = 'Very Good'
data.loc[data['writing score'] >= 90 , 'writing score categ'] = 'Excellent'
```

```python
# Conversion de reading score
data['reading score categ'] = 'Failed'

data.loc[data['reading score'] <= 35, 'reading score categ'] = 'Failed'
data.loc[(data['reading score'] > 35) & (data['reading score'] <= 49), 'reading score categ'] = 'Fair'
data.loc[(data['reading score'] >= 50) & (data['reading score'] <= 64), 'reading score categ'] = 'Passed'
data.loc[(data['reading score'] >= 66) & (data['reading score'] <= 80), 'reading score categ'] = 'Good'
data.loc[(data['reading score'] >= 81) & (data['reading score'] <= 90), 'reading score categ'] = 'Very Good'
data.loc[data['reading score'] >= 90 , 'reading score categ'] = 'Excellent'
```

```python
# Conversion de math score
data['math score categ'] = 'Failed'

data.loc[data['math score'] <= 35, 'math score categ'] = 'Failed'
data.loc[(data['math score'] > 35) & (data['math score'] <= 49), 'math score categ'] = 'Fair'
data.loc[(data['math score'] >= 50) & (data['math score'] <= 64), 'math score categ'] = 'Passed'
data.loc[(data['math score'] >= 66) & (data['math score'] <= 80), 'math score categ'] = 'Good'
data.loc[(data['math score'] >= 81) & (data['math score'] <= 90), 'math score categ'] = 'Very Good'
data.loc[data['math score'] >= 90 , 'math score categ'] = 'Excellent'
```
